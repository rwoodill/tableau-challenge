```python
import pandas as pd
from pathlib import Path
```

```python
load_201306 = Path("change_in_ridership/201306-citibike-tripdata.csv")
june2013_data = pd.read_csv(load_201306)
june2013_data.head()
```

```python
#rename columns
june2013_data.rename(columns={'starttime': 'started_at', 'stoptime': 'ended_at', 'start station id': 'start_station_id', 
                              'start station name': 'start_station_name', 'start station latitude': 'start_lat',
                             'start station longitude': 'start_lng', 'end station id': 'end_station_id', 
                             'end station name': 'end_station_name', 'end station latitude': 'end_lat', 
                             'end station longitude': 'end_lng', 'usertype': 'member_casual'}, inplace=True)
```

```python
#delete unneccessary columns
columns= ['tripduration', 'bikeid', 'birth year', 'gender']
june2013_data.drop(columns, inplace=True, axis=1)
june2013_data.head()
```

```python
#rename member/casual
june2013_data.loc[june2013_data["member_casual"] == "Customer", "member_casual"] = "casual"
june2013_data.loc[june2013_data["member_casual"] == "Subscriber", "member_casual"] = "member"
june2013_data.head()
```

```python
cols = list(june2013_data.columns.values)
cols
```

```python
load_201806 = Path("change_in_ridership/201806-citibike-tripdata.csv")
june2018_data = pd.read_csv(load_201806)
june2018_data.head()
```

```python
#rename columns
june2018_data.rename(columns={'starttime': 'started_at', 'stoptime': 'ended_at', 'start station id': 'start_station_id', 
                              'start station name': 'start_station_name', 'start station latitude': 'start_lat',
                             'start station longitude': 'start_lng', 'end station id': 'end_station_id', 
                             'end station name': 'end_station_name', 'end station latitude': 'end_lat', 
                             'end station longitude': 'end_lng', 'usertype': 'member_casual'}, inplace=True)
```

```python
#delete unneccessary columns
columns= ['tripduration', 'bikeid', 'birth year', 'gender']
june2018_data.drop(columns, inplace=True, axis=1)
june2018_data.head()
```

```python
#rename member/casual
june2018_data.loc[june2018_data["member_casual"] == "Customer", "member_casual"] = "casual"
june2018_data.loc[june2018_data["member_casual"] == "Subscriber", "member_casual"] = "member"
june2018_data.head()
```

```python
cols = list(june2018_data.columns.values)
cols
```

```python
load_202306 = Path("change_in_ridership/202306-citibike-tripdata.csv")
june2023_data = pd.read_csv(load_202306)
```

```python
#delete some extra columns
columns = ['ride_id', 'rideable_type']
june2023_data.drop(columns, inplace=True, axis=1)
june2023_data.head()
```

```python
cols = list(june2023_data.columns.values)
cols
```

```python
june2023_data = june2023_data[['started_at',
                               'ended_at',
                               'start_station_id',
                               'start_station_name',
                               'end_station_id',
                               'end_station_name',
                               'start_lat',
                               'start_lng',
                               'end_lat',
                               'end_lng',
                               'member_casual']]
```

```python
#join them all
df = pd.concat([june2023_data, june2018_data, june2013_data])
df
```

```python
#export 
compression_opts = dict(method='zip',
                        archive_name='out.csv')
df.to_csv('out.zip', index=False,
          compression=compression_opts) 
```
